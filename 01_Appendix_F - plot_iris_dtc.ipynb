## Decision Trees
Decision Trees (DTs) are a non-parametric supervised learning method used for classification and regression. The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features.

```python
%matplotlib inline

print(__doc__)

from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
```

**DecisionTreeClassifier** is a class capable of performing multi-class classification on a dataset.

As with other classifiers, DecisionTreeClassifier takes as input two arrays: an array **X**, sparse or dense, of size [n_samples, n_features] holding the training samples, and an array **Y** of integer values, size [n_samples], holding the class labels for the training samples:

```python
#Import the required classes
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

As with other classifiers, DecisionTreeClassifier takes as input two arrays: an array **X**, sparse or dense, of size [n_samples, n_features] holding the training samples,

```python
# Load data
iris = load_iris()

# On this example the array X == iris.data and y == iris.target
#print(iris.data)
#print(iris.target)
```

After being fitted, the model can then be used to predict the class of samples:

```python
# Fit the classifier to the data
clf = tree.DecisionTreeClassifier()
clf = clf.fit(iris.data, iris.target)
print(clf)
```

Once trained, you can plot the tree with the plot_tree function:

```python
plt.figure(figsize=(50,50))
plot_tree(clf,
          feature_names=iris.feature_names,
          class_names=iris.target_names,
          filled=True,
          rounded=True)

plt.show()
```

```python
df_diabetes = pd.read_csv('./diabetes.csv')
df_diabetes.head()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

X = df_diabetes.iloc[:, :-1]
y = df_diabetes.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)

```

```python
plt.figure(figsize=(50,50))
plot_tree(clf,
          feature_names=list(X_train.columns.values),
          class_names=list(map(str, list(y_train.unique()))),
          filled=True,
          rounded=True)

plt.show()
```
